```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics
import datetime

from statsmodels.tsa.stattools import adfuller

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('reviews.csv', sep=',')
```

```python
# Convert 'date' object type to datetime type
df['date'] = pd.to_datetime(df['date'])
```

```python
# Get counts of listings by date
dfx = df.groupby('date').agg({'ids':'count'})
```

```python
print(dfx.head(), '\n')
print(dfx.tail())
```

```python
# Plot original time series
fig, ax = plt.subplots(1,1)
ax.plot(dfx['ids'], color='#23b5ea')

ax.set_title('Original Time Series of Listing Id Counts Over Time')
plt.xticks(rotation=20)
plt.grid(alpha=0.3)

plt.show()
plt.close()
```

We can see from the time series above, that there is an apparent increasing trend, and some seasonality effects near the last quarter of each year.

Hence, this time series is **NOT Stationary**

```python
def test_stationarity(df, ts):
    
    # Get weekly rolling average and rolling standard deviation of time series 
    rol_avg = df[ts].rolling(window=7, center=False).mean()
    rol_std = df[ts].rolling(window=7, center=False).std()
    
    # Plot original, rolling avg, rolling std
    plt.plot(df[ts], color='#23b5ea', label='Original')
    plt.plot(rol_avg, color='#d81cdb', label = 'Rolling Average')
    plt.plot(rol_std, color='#adadad', label = 'Rolling Std')
    
    plt.legend(loc='best')
    plt.title('Rolling Average & Std: {}'.format(ts))
    plt.xticks(rotation=20)
    plt.grid(alpha=0.3)
    
    plt.show()
    plt.close()
    
    # Apply Dickey Fuller Test on time series
    dftest = adfuller(df[ts], autolag='AIC')
    
    # Format the DF Test output
    df_results = pd.Series(dftest[0:4], index=['Test Statistic',
                                              'p-value',
                                              '# of Lags Used',
                                              '# Observations Used'])
    for key, value in dftest[4].items():
        df_results['Critical Value {}'.format(key)] = value
    
    print(df_results)
```

```python
test_stationarity(dfx, 'ids')
```

```python
def plot_transformation(df, ts, ts_transform):
    fig, ax = plt.subplots(1,1)
    ax.plot(df[ts], color='#23b5ea')
    ax.plot(df[ts_transform], color='#d81cdb')
    
    ax.set_title('{} & {} Time Series'.format(ts, ts_transform))
    ax.legend([ts, ts_transform])
    
    plt.xticks(rotation=20)
    plt.grid(alpha=0.3)
    
    plt.show()
    plt.close()
```

```python
# Apply some transformations to time series data

# ts 7 day moving average
dfx['ids_mov_avg'] = dfx['ids'].rolling(window=7, center=False).mean()

# log(ts)
dfx['ids_log'] = dfx['ids'].apply(lambda x: np.log(x))

# log(ts) 7 day moving average
dfx['ids_log_mov_avg'] = dfx['ids_log'].rolling(window=7, center=False).mean()

# sqrt(ts)
dfx['ids_sqrt'] = dfx['ids'].apply(lambda x: np.sqrt(x))

# sqrt(ts) 7 day moving average
dfx['ids_sqrt_mov_avg'] = dfx['ids_sqrt'].rolling(window=7, center=False).mean()

# 1st order differencing
dfx['ids_1st_diff'] = dfx['ids'].diff()
dfx['ids_1st_diff'] = dfx['ids_1st_diff'].fillna(0)

# 1st order differencing 7 day moving average
dfx['ids_1st_diff_mov_avg'] = dfx['ids_1st_diff'].rolling(window=7, center=False).mean()

# 1st order differencing on log(ts)
dfx['ids_log_diff'] = dfx['ids_log'].diff()
dfx['ids_log_diff'] = dfx['ids_log_diff'].fillna(0)

# 1st order differencing on log(ts) 7 day moving average
dfx['ids_log_diff_mov_avg'] = dfx['ids_log_diff'].rolling(window=7, center=False).mean()
```

```python
dfx.columns
```

```python
# Plot Some Transformations

plot_transformation(dfx, 'ids', 'ids_mov_avg')

plot_transformation(dfx, 'ids_log', 'ids_log_mov_avg')

plot_transformation(dfx, 'ids_sqrt', 'ids_sqrt_mov_avg')

plot_transformation(dfx, 'ids_1st_diff', 'ids_1st_diff_mov_avg')

plot_transformation(dfx, 'ids_log_diff', 'ids_log_diff_mov_avg')
```

```python
# Test stationarity on transformations

test_stationarity(dfx, 'ids')

test_stationarity(dfx, 'ids_log')

test_stationarity(dfx, 'ids_sqrt')

test_stationarity(dfx, 'ids_1st_diff')
```

```python
# Decompose Seasonal & Trend Components of Time Series
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
def plot_decomp(df, ts, trend, seasonal, residual):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(20,7), sharex=True)
    
    ax1.plot(df[ts], color='#23b5ea', label='Original')
    ax1.legend(loc='best')
    ax1.grid(alpha=0.3)
    
    ax2.plot(df[trend], color='#23b5ea', label='Trend')
    ax2.legend(loc='best')
    ax2.grid(alpha=0.3)
    
    ax3.plot(df[seasonal], color='#23b5ea', label='Seasonality')
    ax3.legend(loc='best')
    ax3.grid(alpha=0.3)
    
    ax4.plot(df[residual], color='#23b5ea', label='Residual')
    ax4.legend(loc='best')
    ax4.grid(alpha=0.3)
    
    plt.suptitle('Trend, Seasonality, Residual of {}'.format(ts))
     
    plt.show()
    plt.close()
```

```python
# decompose ids_log

decomp = seasonal_decompose(dfx['ids_log'], freq=365)
dfx['trend'] = decomp.trend
dfx['seasonal'] = decomp.seasonal
dfx['residual'] = decomp.resid

plot_decomp(dfx, 'ids_log', 'trend', 'seasonal', 'residual')

test_stationarity(df = dfx.dropna(), ts ='residual')
```

```python
# decompose ids_log_diff

decomp2 = seasonal_decompose(dfx['ids_log_diff'], freq=365)
dfx['trend2'] = decomp2.trend
dfx['seasonal2'] = decomp2.seasonal
dfx['residual2'] = decomp2.resid

plot_decomp(dfx, 'ids_log_diff', 'trend2', 'seasonal2', 'residual2')

test_stationarity(df = dfx.dropna(), ts = 'ids_log_diff')
```

# ARIMA Modeling On 'ids_log_diff'

```python
# Figure out p, d, q parameters using ACF, PACF plots
```

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
def plot_acf_pacf(df, ts):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
    
    lag_acf = acf(np.array(df[ts]), nlags=20)
    lag_pacf = pacf(np.array(df[ts]), nlags=20)
    
    # Plot ACF
    ax1.plot(lag_acf, color='#23b5ea')
    ax1.grid(alpha=0.3)
    ax1.axhline(y=0,linestyle='--',color='gray')
    ax1.axhline(y=-1.96/np.sqrt(len(df[ts])),linestyle='--',color='gray')
    ax1.axhline(y=1.96/np.sqrt(len(df[ts])),linestyle='--',color='gray')
    ax1.set_title('ACF: {}'.format(ts))
    
    # Plot PACF
    ax2.plot(lag_pacf, color='#23b5ea')
    ax2.grid(alpha=0.3)
    ax2.axhline(y=0,linestyle='--',color='gray')
    ax2.axhline(y=-1.96/np.sqrt(len(df[ts])),linestyle='--',color='gray')
    ax2.axhline(y=1.96/np.sqrt(len(df[ts])),linestyle='--',color='gray')
    ax2.set_title('PACF: {}'.format(ts))
    
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
plot_acf_pacf(dfx, 'ids_log_diff')
```

```python
# Implement ARIMA Model
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
def arima(df, ts, p, d, q):
    
    # instantiate arima model
    model = ARIMA(df[ts], order=(p,d,q))
    
    # fit
    fit_model = model.fit(disp=-1)
    
    len_model = len(fit_model.fittedvalues)
    ts_modified = df[ts][-len_model:]
    
    # sum squared error / residual sum square
    rss = sum((fit_model.fittedvalues - ts_modified)**2)
    
    # root mean square error
    rmse = np.sqrt(rss / len(df[ts]))
    
    # plot fit
    plt.plot(df[ts], color='#23b5ea')
    plt.plot(fit_model.fittedvalues, color='#d81cdb')
    
    plt.title('{} ARIMA({},{},{})'.format(ts, p, d, q))
    
    plt.show()
    plt.close()
    
    return fit_model
```
